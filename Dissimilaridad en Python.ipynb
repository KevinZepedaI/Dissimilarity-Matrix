```python
import numpy as np
import pandas as pd


array1 = np.array([0, 0, 0, .5])
len_array1= len(array1)

array2 =array1
len_array2=(len(array2))

M = [ ]

for i in array1:
  for j in array2:
    M.append(np.absolute(i-j))


Mn = np.reshape(M, (len_array1,len_array2))
print('La matrix de disimilaridad es la siguiente: ', Mn, sep='\n')
print(' ', sep='\n')

print('El promedio de las distancias es: ', Mn.mean())
print(' ', sep='\n')

Mnn = np.max(Mn)
print('El valor más disímil es: ', Mnn)
print(' ', sep='\n')




result = np.where(Mn == np.amax(Mn))

print('Las coordenadas del valor más disimil son (renglón, columna, iniciando desde 0, no de 1): ', sep='\n')
listOfCordinates = list(zip(result[0], result[1]))
for cord in listOfCordinates:
        print(cord)

```
